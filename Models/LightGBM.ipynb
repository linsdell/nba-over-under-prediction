```
import datetime
import lightgbm as lgb
import numpy as np
import os
import pandas as pd
import random
from tqdm import tqdm
from sklearn.model_selection import train_test_split
#import haversine

# Importing the dataset
dataset = pd.read_csv('nba_data_v1.csv')
#dataset.head()
# predicting the total score
y = dataset["Total.Score"]
X = dataset.iloc[:, 4:227]
X_training = X.drop(columns=['Line'])

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_training, y, test_size = 0.25)

params = {
    'boosting_type' : "gbdt",
    'objective': 'regression',
    'max_depth' : 15,
    'num_leaves': 100,
    'min_data_in_leaf':100,
    'colsample_bytree':.5,
    'n_estimators':400,
    'min_child_weight':5,
    'min_child_samples':10,
    'subsample':.632, # Standard RF bagging fraction
    'subsample_freq':1,
    'min_split_gain':0,
    'reg_alpha':0, # L2 regularization
    'reg_lambda':5,
    'n_jobs':3
    }
d_train = lgb.Dataset(x_train, label=y_train)
gbm = lgb.train(params, d_train, num_boost_round=100)


predictions = gbm.predict(x_test)
win = 0
loss = 0 
try:
  for index, instance in dataset.iterrows():
   actual = instance["Total.Score"]
   line = instance["Line"]
   prediction = round(predictions[index])
   print((index+1), "actual = ", actual,"prediction = ", prediction, "Line  = ", line)
   if prediction > line:
    if actual > line:
     win = win + 1
    elif actual <= line:
     loss = loss + 1
   elif prediction <= line: 
    if actual < line: 
     win = win + 1
    elif actual >= line:
     loss = loss + 1   
   #print(win,loss)  
except IndexError:
  print("")

total = (win)/(win+loss)
print("Total win percentage is ", total*100, "%")
```
